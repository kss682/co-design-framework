```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
states = pd.read_csv("build/states.csv")
states.head()
```

```python
plt.subplot(2, 1, 1)
plt.plot(states['# t'], states['x'])
plt.title('Position vs Time for 5G Delay Distribution')
plt.xlabel('Time')
plt.ylabel('Position')
```

```python
plt.subplot(2, 1, 2)
plt.plot(states['# t'], states['phi'])
plt.title('Pendulum Angle vs Time for 5G Delay Distribution')
plt.xlabel('Time')
plt.ylabel('Angle')
```
